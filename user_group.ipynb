## 权限管理
1. 模型和所属关系
    + mode:rwxrwxrwx
    + ownership:user,group
2. ls -l
    + rwxrwxrwx
    + user,group,other
3. chmod 
    + u g o a
    + -R 递归修改目录
    + chmod g=rw,u=r xx 
    + chmod ug=rw xx
    + chmod u-x xx
    + chmod 775 xx
    + chmod --reference xxx xxx

4. chown
    + -R
    + chown -R user:group xxx/
5. chgrp
    + chgrp group /xxx
6. 权限
    + 文件：
        + r：可获取文件数据
        + w：可修改文件数据
        + x：可将此文件运行为进程
    + 目录：
        + r：可使用ls命令获取其下所有文件列表
        + w：可修改此目录下的文件列表，即创建删除文件
        + x：可cd至此目录，可使用ls -l来获取文件的详细信息
7. 权限组合模型
    + --- 000 0
    + --x 001 1
    + -w- 010 2
    + -wx 011 3
    + r-- 100 4
    + r-x 101 5
    + rw- 110 6
    + rwx 111 7
